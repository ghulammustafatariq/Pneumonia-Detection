```python
import os
import time

# --- 1. Paths ko define karein ---
DRIVE_IMAGE_DIR = "/content/drive/MyDrive/Pneumonia_Advanced_Project/2_processed_data/images"
DRIVE_PROJECT_DIR = "/content/drive/MyDrive/Pneumonia_Advanced_Project"

# Local folder jahan images copy hon gi
LOCAL_IMAGE_DIR = "/content/local_images"

TAR_FILENAME = "images_archive.tar.gz"
LOCAL_TAR_PATH = f"/content/{TAR_FILENAME}"
DRIVE_TAR_PATH = os.path.join(DRIVE_PROJECT_DIR, TAR_FILENAME)

os.makedirs(LOCAL_IMAGE_DIR, exist_ok=True)
print("--- Naya Tareeqa Shuru (Plan D - With Progress Counter) ---")

# --- STEP 1: Drive se Local Disk par Copy (With Progress) (10-15 min) ---
print(f"\nSTEP 1/3: 30,000 files ko Drive se local disk par copy kiya ja raha hai...")
print("Aapko har file ka progress neeche dikhega.")
start_time = time.time()

# --- YEH HAI ASAL CHANGE (rsync -aP) ---
# -a = archive mode (recursive)
# -P = --progress (har file ka progress dikhaye) aur --partial (resume kar sake)
# / (slash) zaroori hai: iska matlab hai "folder ke andar ka content"
!rsync -aP "{DRIVE_IMAGE_DIR}/" "{LOCAL_IMAGE_DIR}"

print(f"Step 1 complete in {(time.time() - start_time)/60:.2f} minutes.")
print(f"Ab '/content/local_images' folder local disk par maujood hai.")

# --- STEP 2: Local Disk par Archive Banana (Bohat Tez: 2-3 min) ---
print(f"\nSTEP 2/3: Local disk par '{TAR_FILENAME}' banayi ja rahi hai...")
start_time = time.time()
# Hum naye local path se archive banayein ge
!tar -czvf "{LOCAL_TAR_PATH}" -C "{LOCAL_IMAGE_DIR}" .
print(f"Step 2 complete in {(time.time() - start_time)/60:.2f} minutes.")

# --- STEP 3: 1 File ko Drive par Waapas Bhejna (5-10 min) ---
print(f"\nSTEP 3/3: 1 archive file ko Google Drive par copy kiya ja raha hai...")
start_time = time.time()
!cp "{LOCAL_TAR_PATH}" "{DRIVE_TAR_PATH}"
print(f"Step 3 complete in {(time.time() - start_time)/60:.2f} minutes.")

print(f"\n--- MUKAMMAL HO GAYA! ---")
print(f"Aapki file '{DRIVE_TAR_PATH}' ab Drive mein save ho gayi hai.")
```

```python
import os

# --- 1. Paths ---
#  In paths ko ehtiyat se check karein
BASE_DIR = "/content/drive/MyDrive/Pneumonia_Advanced_Project/2_processed_data"
SOURCE_DIR = os.path.join(BASE_DIR, "images")
DEST_DIR = os.path.join(BASE_DIR, "images_part_2") # Naye folder ka naam

# --- 2. Naya folder banayein ---
print(f"Naya folder banaya ja raha hai: {DEST_DIR}")
# Use {VAR} syntax to pass Python variables to the shell
!mkdir -p {DEST_DIR}

# --- 3. 10,000 files ko move karein (Corrected Syntax) ---
print(f"\n10,000 files ko '{SOURCE_DIR}' se '{DEST_DIR}' mein MOVE kiya ja raha hai...")

# Code:
# Hum {VAR} istemaal kar rahe hain (e.g. {SOURCE_DIR})
# Hum xargs ke liye placeholder badal kar '_FILE_' kar rahe hain taake { } ka conflict na ho
!ls -1 {SOURCE_DIR} | head -n 10000 | xargs -I _FILE_ mv "{SOURCE_DIR}/_FILE_" "{DEST_DIR}/"

print(f"\n--- Move complete ---")
print(f"Check karein ke files '{DEST_DIR}' mein move ho gayi hain.")
```
