```python
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense, Flatten, Dropout
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.layers import RandomFlip, RandomRotation, RandomZoom, RandomTranslation
from tensorflow.keras.applications.vgg16 import VGG16
import numpy as np
import math
import matplotlib.pyplot as plt
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

# --- 1. Configuration and Paths ---
IMAGE_SIZE = (224, 224)
BATCH_SIZE = 32
INITIAL_EPOCHS = 20
FINE_TUNE_EPOCHS = 10
AUTOTUNE = tf.data.AUTOTUNE

train_path = '/content/drive/MyDrive/data/chest_xray/train'
test_path = '/content/drive/MyDrive/data/chest_xray/test'
# We will not use the old 'val_path'


# --- 2. Build High-Performance tf.data Pipelines (THE 80/20 SPLIT FIX) ---
print("Loading and splitting training data for validation...")
# This will use 80% (around 4180 images) for training
train_ds = tf.keras.utils.image_dataset_from_directory(
    train_path,
    labels='inferred',
    label_mode='binary',
    image_size=IMAGE_SIZE,
    interpolation='nearest',
    batch_size=BATCH_SIZE,
    shuffle=True,
    validation_split=0.2,  # <--- CREATE A 20% VALIDATION SPLIT
    subset="training",     # <--- MARK THIS AS THE TRAINING SUBSET
    seed=123 # Add a seed for reproducible splits
)

# This will use 20% (around 1044 images) for validation
val_ds = tf.keras.utils.image_dataset_from_directory(
    train_path,            # <--- LOAD FROM THE SAME 'train_path'
    labels='inferred',
    label_mode='binary',
    image_size=IMAGE_SIZE,
    interpolation='nearest',
    batch_size=BATCH_SIZE,
    shuffle=False,
    validation_split=0.2,  # <--- USE THE SAME 20% SPLIT
    subset="validation",   # <--- MARK THIS AS THE VALIDATION SUBSET
    seed=123 # Add a seed for reproducible splits
)

# The test dataset remains the same
print("Loading test data...")
test_ds = tf.keras.utils.image_dataset_from_directory(
    test_path,
    labels='inferred',
    label_mode='binary',
    image_size=IMAGE_SIZE,
    interpolation='nearest',
    batch_size=BATCH_SIZE,
    shuffle=False
)

# --- 3. Preprocessing Function (THE 'TypeError' FIX) ---
def normalize_image(image, label):
    """Casts image to float and rescales to [0, 1]."""
    image = tf.cast(image, tf.float32) / 255.0
    return image, label

# --- 4. Optimize the Pipelines (THE SPEED FIX) ---
print("Building and optimizing data pipelines...")
train_ds = (
    train_ds
    .map(normalize_image, num_parallel_calls=AUTOTUNE)
    .cache()
    .prefetch(buffer_size=AUTOTUNE)
)

val_ds = (
    val_ds
    .map(normalize_image, num_parallel_calls=AUTOTUNE)
    .cache()
    .prefetch(buffer_size=AUTOTUNE)
)

test_ds = (
    test_ds
    .map(normalize_image, num_parallel_calls=AUTOTUNE)
    .cache()
    .prefetch(buffer_size=AUTOTUNE)
)
print("Data pipelines built.")


# --- 5. Calculate Class Weights (THE 11-MINUTE WAIT FIX) ---
print("Using pre-calculated class weights for 80% training split...")

# These are the correct numbers for the 80% split (4180 images)
total_samples = 4180
normal_count = 1073
pneumonia_count = 3107

print(f"New Training Set: Total: {total_samples}, Normal: {normal_count}, Pneumonia: {pneumonia_count}")

# Instant calculation
weight_for_normal = total_samples / (2.0 * normal_count)
weight_for_pneumonia = total_samples / (2.0 * pneumonia_count)
class_weights = {0: weight_for_normal, 1: weight_for_pneumonia}
print(f"Calculated Class Weights: {class_weights}")


# --- 6. Model Setup (Phase 1: Frozen) (THE SPEED FIX) ---
# GPU-powered augmentation
data_augmentation = Sequential([
    RandomFlip("horizontal"),
    RandomRotation(0.2),
    RandomZoom(0.2),
    RandomTranslation(height_factor=0.2, width_factor=0.2),
], name="data_augmentation")

inputs = Input(shape=IMAGE_SIZE + (3,))
x = data_augmentation(inputs) # Apply augmentation on GPU

vgg_base = VGG16(input_shape=IMAGE_SIZE + (3,), weights='imagenet', include_top=False, name="vgg16")
vgg_base.trainable = False # Freeze the base

x = vgg_base(x, training=False)
x = Flatten()(x)
x = Dense(128, activation='relu')(x)
x = Dropout(0.5)(x)
prediction = Dense(1, activation='sigmoid')(x)
model = Model(inputs=inputs, outputs=prediction)


# --- 7. Compilation and Training (Phase 1) ---
print("\n--- PHASE 1: Training the Head ---")
callbacks = [
    ModelCheckpoint('best_model_vgg16_phase1.h5', monitor='val_loss', save_best_only=True, verbose=1),
    EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True, verbose=1)
]

model.compile(
    loss='binary_crossentropy',
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
    metrics=['accuracy', tf.keras.metrics.Precision(), tf.keras.metrics.Recall()]
)
model.summary()

history = model.fit(
    train_ds,
    validation_data=val_ds, # Now using the reliable 20% split
    epochs=INITIAL_EPOCHS,
    callbacks=callbacks,
    class_weight=class_weights
)
history_phase1 = history.history
print("\n--- Phase 1 Training Complete ---")


# --- 8. Phase 2: Fine-Tuning (THE 'ValueError' CRASH FIX) ---
print("\n--- PHASE 2: Fine-Tuning the Top Layers ---")
# 'model' already has the best weights from Phase 1 restored by EarlyStopping

# Get the VGG16 base layer from our model by its name
vgg_base = model.get_layer("vgg16")

# Un-freeze the base model
vgg_base.trainable = True

# Freeze all layers *except* the last block
fine_tune_at = 15
for layer in vgg_base.layers[:fine_tune_at]:
    layer.trainable = False

print(f"Unfreezing from layer {vgg_base.layers[fine_tune_at].name} onwards...")

# Re-compile the *same model* with a low learning rate
model.compile(
    loss='binary_crossentropy',
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-5), # 10x smaller LR
    metrics=['accuracy', tf.keras.metrics.Precision(), tf.keras.metrics.Recall()]
)
model.summary() # Note the change in trainable parameters

# Define new callbacks for the fine-tuned model
callbacks_ft = [
    ModelCheckpoint('best_model_vgg16_finetuned.h5', monitor='val_loss', save_best_only=True, verbose=1),
    EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True, verbose=1)
]

# Continue training the *same model*
history_ft = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=INITIAL_EPOCHS + FINE_TUNE_EPOCHS,
    initial_epoch=history.epoch[-1] + 1, # Start from where we left off
    callbacks=callbacks_ft,
    class_weight=class_weights
)

# --- 9. Final Evaluation and Visualization ---
print("\n--- Evaluating Final Fine-Tuned Model on Test Set ---")
# Load the best fine-tuned model (saved by the callback) for final eval
final_model = tf.keras.models.load_model('best_model_vgg16_finetuned.h5')

test_loss, test_accuracy, test_precision, test_recall = final_model.evaluate(test_ds)
print(f"Test Accuracy: {test_accuracy:.4f}")
print(f"Test Loss: {test_loss:.4f}")
print(f"Test Precision: {test_precision:.4f}")
print(f"Test Recall: {test_recall:.4f}")

# Plotting
acc = history_phase1['accuracy'] + history_ft.history['accuracy']
val_acc = history_phase1['val_accuracy'] + history_ft.history['val_accuracy']
loss = history_phase1['loss'] + history_ft.history['loss']
val_loss = history_phase1['val_loss'] + history_ft.history['val_loss']

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(acc)
plt.plot(val_acc)
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')

plt.subplot(1, 2, 2)
plt.plot(loss)
plt.plot(val_loss)
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')

plt.tight_layout()
plt.show()

print("Training and evaluation complete. Best model saved as 'best_model_vgg16_finetuned.h5'")
```

```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report

# --- 1. Configuration ---
MODEL_PATH = '/content/drive/MyDrive/data/best_model_vgg16_finetuned (1).h5'
TRAIN_PATH = '/content/drive/MyDrive/data/chest_xray/train' # Path to the 'train' folder
IMAGE_SIZE = (224, 224)
BATCH_SIZE = 32
AUTOTUNE = tf.data.AUTOTUNE

# --- 2. Load the Saved Model ---
print(f"Loading model from {MODEL_PATH}...")
model = load_model(MODEL_PATH)
print("Model loaded successfully.")

# --- 3. Preprocessing Function (Must be IDENTICAL to training) ---
def normalize_image(image, label):
    """Casts image to float and rescales to [0, 1]."""
    image = tf.cast(image, tf.float32) / 255.0
    return image, label

# --- 4. Load the Training Dataset ---
# IMPORTANT: We set 'shuffle=False' to ensure predictions and labels line up.
# We are using the *entire* train folder, not just the 80% split.
print("Loading training data for evaluation (this may take a moment)...")
train_ds = tf.keras.utils.image_dataset_from_directory(
    TRAIN_PATH,
    labels='inferred',
    label_mode='binary',
    image_size=IMAGE_SIZE,
    interpolation='nearest',
    batch_size=BATCH_SIZE,
    shuffle=False  # <-- CRITICAL: Must be False for confusion matrix
)

# Optimize the dataset for prediction
train_ds = (
    train_ds
    .map(normalize_image, num_parallel_calls=AUTOTUNE)
    .cache()
    .prefetch(buffer_size=AUTOTUNE)
)
print("Data loaded and preprocessed.")

# --- 5. Get Predictions and True Labels ---
print("Generating predictions on training data...")
# Get all the raw probability predictions
y_pred_probs = model.predict(train_ds)

# Get all the true labels
y_true = []
for images, labels in train_ds:
    y_true.extend(labels.numpy())

# Convert arrays to numpy
y_true = np.array(y_true)
# Squeeze the array if it's (N, 1) to just (N,)
y_true = y_true.squeeze()

# Convert probabilities (e.g., 0.98) to binary class labels (1)
y_pred = (y_pred_probs > 0.5).astype(int).squeeze()

# --- 6. Generate and Print Metrics ---
print("\n--- Classification Report (Training Set) ---")
# (Assuming 0 = NORMAL, 1 = PNEUMONIA)
print(classification_report(y_true, y_pred, target_names=['NORMAL (0)', 'PNEUMONIA (1)']))


print("\n--- Confusion Matrix (Training Set) ---")
# Generate the matrix
cm = confusion_matrix(y_true, y_pred)

# Plot the matrix
plt.figure(figsize=(7, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='rocket',
            xticklabels=['Predicted: NORMAL', 'Predicted: PNEUMONIA'],
            yticklabels=['Actual: NORMAL', 'Actual: PNEUMONIA'],
            linewidths=0.5)
plt.title('Confusion Matrix (on Training Data)', fontsize=14)
plt.show()
```
